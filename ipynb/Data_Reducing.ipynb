```python
import pandas as pd

df = pd.read_csv("Data/cicids2017_cleaned.csv")
df['Attack Type'].value_counts()
```

```python
sample_sizes = {
    'Normal Traffic': 60000,
    'DoS': 30000,
    'DDoS': 25000,
    'Port Scanning': 25000,
    'Brute Force': 15000,
    'Web Attacks': 10000,
    'Bots': 6000   # moderate upsampling
}
```

```python
balanced = pd.DataFrame()
```

```python
for attack, target_size in sample_sizes.items():
    subset = df[df['Attack Type'] == attack]
    if len(subset) > target_size:
        subset = subset.sample(n=target_size, random_state=42)
    else:
        subset = subset.sample(n=target_size, replace=True, random_state=42)
    balanced = pd.concat([balanced, subset])

balanced.to_csv("CICIDS_balanced_171k.csv", index=False)
```
