```python
import pandas as pd

# Load new dataset
new_df = pd.read_csv("Data/NewTestData.csv")

# 1️⃣ Remove leading/trailing spaces from column names
new_df.columns = new_df.columns.str.strip()

# 2️⃣ Define the columns used during training (your original df)
required_cols = [
    "Destination Port", "Flow Duration", "Total Fwd Packets", "Total Length of Fwd Packets",
    "Fwd Packet Length Max", "Fwd Packet Length Min", "Fwd Packet Length Mean", "Fwd Packet Length Std",
    "Bwd Packet Length Max", "Bwd Packet Length Min", "Bwd Packet Length Mean", "Bwd Packet Length Std",
    "Flow Bytes/s", "Flow Packets/s", "Flow IAT Mean", "Flow IAT Std", "Flow IAT Max", "Flow IAT Min",
    "Fwd IAT Total", "Fwd IAT Mean", "Fwd IAT Std", "Fwd IAT Max", "Fwd IAT Min",
    "Bwd IAT Total", "Bwd IAT Mean", "Bwd IAT Std", "Bwd IAT Max", "Bwd IAT Min",
    "Fwd Header Length", "Bwd Header Length", "Fwd Packets/s", "Bwd Packets/s",
    "Min Packet Length", "Max Packet Length", "Packet Length Mean", "Packet Length Std", "Packet Length Variance",
    "FIN Flag Count", "PSH Flag Count", "ACK Flag Count", "Average Packet Size",
    "Subflow Fwd Bytes", "Init_Win_bytes_forward", "Init_Win_bytes_backward", "act_data_pkt_fwd",
    "min_seg_size_forward", "Active Mean", "Active Max", "Active Min", "Idle Mean", "Idle Max", "Idle Min" ,"Label"
]

# 3️⃣ Drop columns not in the trained model
extra_cols = set(new_df.columns) - set(required_cols)
if extra_cols:
    print(f"Dropping {len(extra_cols)} extra columns: {extra_cols}")
    new_df = new_df.drop(columns=list(extra_cols))

# 4️⃣ (Optional) Check for missing columns
missing_cols = set(required_cols) - set(new_df.columns)
if missing_cols:
    print(f"⚠️ Missing columns: {missing_cols}")
else:
    print("✅ All required columns are present.")

# 5️⃣ Keep the exact column order as training data
new_df = new_df[required_cols]

print(f"Final new_df shape: {new_df.shape}")
```
